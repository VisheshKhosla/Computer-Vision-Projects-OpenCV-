```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("opencv-logo.png")
img=cv2.imread('b-Noisy-frame-465-of-moving-cars-corrupted-with-Salt-and-Pepper-noise-under-50-noise.png') # For Median Blur
img=cv2.imread('lena.jpg') # For preserving the edges in case of bilateral filter
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel) # -1 is the depth of the kernel
blur=cv2.blur(img,(5,5)) # Output pixel is the average of neighboring pixels of the kernel 
g_blur=cv2.GaussianBlur(img,(5,5),0) # The centre pixels have more weight and 0 is sigmaX
m_blur=cv2.medianBlur(img,5) # Removing salt and pepper noise or removing white and black dots
b_blur=cv2.bilateralFilter(img,9,75,75) # 9 is the diameter of the neighboring pixels
titles=['Images','2DFilter','Blur','Gaussian_Blur','Median_Blur','Bilateral Filter']
images=[img,dst,blur,g_blur,m_blur,b_blur]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
```
