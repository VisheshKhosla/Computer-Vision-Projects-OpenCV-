```python
import numpy as np
import cv2
```

```python
def  nothing(x):
    print(x)
```

```python
# Creating a trackbar
cv2.namedWindow("Tracking")
cv2.createTrackbar("LH","Tracking",0,255,nothing) # This is for lower hue
cv2.createTrackbar("LS","Tracking",0,255,nothing)
cv2.createTrackbar("LV","Tracking",0,255,nothing)
cv2.createTrackbar("UH","Tracking",255,255,nothing)
cv2.createTrackbar("US","Tracking",255,255,nothing)
cv2.createTrackbar("UV","Tracking",255,255,nothing)
# HSV values are also between 0 to 255
```

```python
# Providing static values to the upper and lower bound
while(1):
    frame=cv2.imread('smarties.png')
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    l_b=np.array([110,50,50]) # This is the lower bound of blue color in hsv
    u_b=np.array([130,255,255])
    
    mask=cv2.inRange(hsv,l_b,u_b) # This mask is only going to cover those pixels with in the range give
    res=cv2.bitwise_and(frame,frame,mask=mask) # Now i am putting the mask on the frame
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("res",res)
    if cv2.waitKey(1)==27:
        break
cv2.destroyAllWindows()
```

```python
# Creating a trackbar
cv2.namedWindow("Tracking")
cv2.createTrackbar("LH","Tracking",0,255,nothing) # This is for lower hue
cv2.createTrackbar("LS","Tracking",0,255,nothing)
cv2.createTrackbar("LV","Tracking",0,255,nothing)
cv2.createTrackbar("UH","Tracking",255,255,nothing)
cv2.createTrackbar("US","Tracking",255,255,nothing)
cv2.createTrackbar("UV","Tracking",255,255,nothing)
# HSV values are also between 0 to 255
# Dynamically changing upper and lower HSV bounds using Trackbar
while(1):
    frame=cv2.imread('photo-1563738068154-8d2e9f19ed62.jfif')
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    l_h=cv2.getTrackbarPos('LH','Tracking')
    l_s=cv2.getTrackbarPos('LS','Tracking')
    l_v=cv2.getTrackbarPos('LV','Tracking')
    u_h=cv2.getTrackbarPos('UH','Tracking')
    u_s=cv2.getTrackbarPos('US','Tracking')
    u_v=cv2.getTrackbarPos('UV','Tracking')
    l_b=np.array([l_h,l_s,l_v]) # This is the lower bound of blue color in hsv
    u_b=np.array([u_h,u_s,u_v]) # Upper bound
    mask=cv2.inRange(hsv,l_b,u_b) # This mask is only going to cover those pixels with in the range give
    res=cv2.bitwise_and(frame,frame,mask=mask) # Now i am putting the mask on the frame
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("res",res)
    if cv2.waitKey(1)==27:
        break
cv2.destroyAllWindows()
```

```python
# Object detection in a video
# Creating a trackbar
cap=cv2.VideoCapture(0)
cv2.namedWindow("Tracking")
cv2.createTrackbar("LH","Tracking",0,255,nothing) # This is for lower hue
cv2.createTrackbar("LS","Tracking",0,255,nothing)
cv2.createTrackbar("LV","Tracking",0,255,nothing)
cv2.createTrackbar("UH","Tracking",255,255,nothing)
cv2.createTrackbar("US","Tracking",255,255,nothing)
cv2.createTrackbar("UV","Tracking",255,255,nothing)
# HSV values are also between 0 to 255
# Dynamically changing upper and lower HSV bounds using Trackbar
while(1):
    _,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    l_h=cv2.getTrackbarPos('LH','Tracking')
    l_s=cv2.getTrackbarPos('LS','Tracking')
    l_v=cv2.getTrackbarPos('LV','Tracking')
    u_h=cv2.getTrackbarPos('UH','Tracking')
    u_s=cv2.getTrackbarPos('US','Tracking')
    u_v=cv2.getTrackbarPos('UV','Tracking')
    l_b=np.array([l_h,l_s,l_v]) # This is the lower bound of blue color in hsv
    u_b=np.array([u_h,u_s,u_v]) # Upper bound
    mask=cv2.inRange(hsv,l_b,u_b) # This mask is only going to cover those pixels with in the range give
    res=cv2.bitwise_and(frame,frame,mask=mask) # Now i am putting the mask on the frame
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("res",res)
    if cv2.waitKey(1)==27:
        break
cap.release()
cv2.destroyAllWindows()
```
